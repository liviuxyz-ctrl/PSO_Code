```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from pyswarms.single import GlobalBestPSO, LocalBestPSO
from pyswarms.utils.functions.single_obj import sphere, rastrigin, ackley, rosenbrock
import optuna
import pandas as pd

# Step 1: Define benchmark functions
functions = {
    "Sphere": sphere,
    "Rastrigin": rastrigin,
    "Ackley": ackley,
    "Rosenbrock": rosenbrock,
}

# Corrected bounds for all benchmark functions
bounds = {
    "Sphere": (np.array([-10] * 30), np.array([10] * 30)),
    "Rastrigin": (np.array([-5.12] * 30), np.array([5.12] * 30)),
    "Ackley": (np.array([-32] * 30), np.array([32] * 30)),
    "Rosenbrock": (np.array([-5] * 30), np.array([10] * 30)),
}

# Default (Non-Optimized) Hyperparameters
default_gbest = {"c1": 0.5, "c2": 0.3, "w": 0.9}
default_lbest = {"c1": 0.5, "c2": 0.3, "w": 0.9, "k": 5, "p": 2}

# Step 2: Define optimization functions
def optimize_gbest(trial, func, bounds):
    c1 = trial.suggest_float("c1", 0.1, 2.0)
    c2 = trial.suggest_float("c2", 0.1, 2.0)
    w = trial.suggest_float("w", 0.1, 1.0)
    options = {"c1": c1, "c2": c2, "w": w}
    optimizer = GlobalBestPSO(n_particles=100, dimensions=30, options=options, bounds=bounds)
    cost, _ = optimizer.optimize(func, iters=200, verbose=False)
    return cost

def optimize_lbest(trial, func, bounds):
    c1 = trial.suggest_float("c1", 0.1, 2.0)
    c2 = trial.suggest_float("c2", 0.1, 2.0)
    w = trial.suggest_float("w", 0.1, 1.0)
    k = trial.suggest_int("k", 2, 10)
    options = {"c1": c1, "c2": c2, "w": w, "k": k, "p": 2}
    optimizer = LocalBestPSO(n_particles=100, dimensions=30, options=options, bounds=bounds)
    cost, _ = optimizer.optimize(func, iters=200, verbose=False)
    return cost

# Step 3: Run hyperparameter optimization
optimized_params = {}
for func_name, func in functions.items():
    print(f"Optimizing parameters for {func_name}...")

    # Optimize Gbest
    study_gbest = optuna.create_study(direction="minimize")
    study_gbest.optimize(lambda trial: optimize_gbest(trial, func, bounds[func_name]), n_trials=10)
    optimized_params[f"{func_name}_Gbest"] = study_gbest.best_params

    # Optimize Lbest
    study_lbest = optuna.create_study(direction="minimize")
    study_lbest.optimize(lambda trial: optimize_lbest(trial, func, bounds[func_name]), n_trials=10)
    optimized_params[f"{func_name}_Lbest"] = study_lbest.best_params

# Step 4: Compare Non-Optimized vs Optimized Results
results = {}
for func_name, func in functions.items():
    print(f"Running comparison for {func_name} function...")

    # Non-Optimized Gbest
    gbest_default = GlobalBestPSO(n_particles=100, dimensions=30, options=default_gbest, bounds=bounds[func_name])
    gbest_default_cost, _ = gbest_default.optimize(func, iters=200, verbose=False)

    # Optimized Gbest
    gbest_opt = GlobalBestPSO(n_particles=100, dimensions=30, options=optimized_params[f"{func_name}_Gbest"], bounds=bounds[func_name])
    gbest_opt_cost, _ = gbest_opt.optimize(func, iters=200, verbose=False)

    # Non-Optimized Lbest
    lbest_default = LocalBestPSO(n_particles=100, dimensions=30, options=default_lbest, bounds=bounds[func_name])
    lbest_default_cost, _ = lbest_default.optimize(func, iters=200, verbose=False)

    # Optimized Lbest
    lbest_opt_params = optimized_params[f"{func_name}_Lbest"]
    lbest_opt_params["p"] = 2
    lbest_opt = LocalBestPSO(n_particles=100, dimensions=30, options=lbest_opt_params, bounds=bounds[func_name])
    lbest_opt_cost, _ = lbest_opt.optimize(func, iters=200, verbose=False)

    # Store results
    results[func_name] = {
        "Gbest Default": gbest_default.cost_history,
        "Gbest Optimized": gbest_opt.cost_history,
        "Lbest Default": lbest_default.cost_history,
        "Lbest Optimized": lbest_opt.cost_history,
    }

# Step 5: Plot Results
def plot_results(results):
    for func_name, data in results.items():
        plt.figure(figsize=(10, 6))
        plt.plot(data["Gbest Default"], label="Gbest Default", linestyle='-', color='blue')
        plt.plot(data["Gbest Optimized"], label="Gbest Optimized", linestyle='--', color='cyan')
        plt.plot(data["Lbest Default"], label="Lbest Default", linestyle='-', color='green')
        plt.plot(data["Lbest Optimized"], label="Lbest Optimized", linestyle='--', color='lime')
        plt.title(f"{func_name} Function: Hyperparameter Comparison")
        plt.xlabel("Iterations")
        plt.ylabel("Cost")
        plt.legend()
        plt.grid()
        plt.tight_layout()
        plt.show()

plot_results(results)

# Step 6: Generate Analysis Table
final_costs = {
    "Function": [],
    "Gbest Default": [],
    "Gbest Optimized": [],
    "Lbest Default": [],
    "Lbest Optimized": [],
}

for func_name, data in results.items():
    final_costs["Function"].append(func_name)
    final_costs["Gbest Default"].append(data["Gbest Default"][-1])
    final_costs["Gbest Optimized"].append(data["Gbest Optimized"][-1])
    final_costs["Lbest Default"].append(data["Lbest Default"][-1])
    final_costs["Lbest Optimized"].append(data["Lbest Optimized"][-1])

df = pd.DataFrame(final_costs)
print(df)
```
